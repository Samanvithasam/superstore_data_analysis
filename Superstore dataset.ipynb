```python
plt.figure(figsize=(9,6))
sns.kdeplot(df['Sales'],color='Orange',label='Sales',fill=True,bw_adjust=.2)
sns.kdeplot(df['Profit'],color='Blue',label='Profit',fill=True,bw_adjust=.2)
plt.xlim([-100,1000])
plt.legend()
```

```python


fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(10,7))

sns.countplot(df,x='Category',ax=axs[0,0])
sns.countplot(df,x='Segment',ax=axs[0,1])
sns.countplot(df,x='Ship Mode',ax=axs[1,0])
sns.countplot(df,x='Region',ax=axs[1,1])
axs[0,0].set_title('Category',fontsize=20)
axs[0,1].set_title('Segment',fontsize=20)
axs[1,0].set_title('Ship Mode',fontsize=20)
axs[1,1].set_title('Region',fontsize=20)


plt.tight_layout()

```

```python
df_segment= df.groupby(['Segment'])[['Sales', 'Discount', 'Profit']].mean()
df_segment
```

```python
df_segment.plot.pie(subplots=True, 
                    autopct='%1.1f%%',
                    figsize=(18, 20),
                    startangle=90,    
                    shadow=True,
                    labels = df_segment.index)
plt.title('Segment wise analysis of Sale, Discount, profit')
```
